```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('titanic')[['pclass', 'age', 'fare', 'survived']]
df
```

```python
df['age'].fillna(df.age.median(),inplace=True)
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X = df[['pclass', 'age', 'fare']]
y = df['survived']
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.33, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled # fit and scaling the whole data
```

```python
# to overcode over fitting
X_train_scaled = scaler.fit_transform(X_train) # fit transformed to train data
X_test_scaled = scaler.transform(X_test) # just fit on the test data
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
classification.fit(X_train_scaled,y_train)
classification.predict(X_test_scaled)
```

```python
# Calculate the mean accuracy on the test data and labels
accuracy = classification.score(X_test_scaled, y_test)
print("Accuracy:", accuracy)
```
